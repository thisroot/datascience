#### datasets.make_circles
генерация "красивого датасета" 

```python
circles = datasets.make_circles()
```

Выводит 2 элмента в типе typle:
- данные (features)
- метки классов (target)

```python
print "features: {}".format(circles[0][:10])
print "target: {}".format(circles[1][:10])
```

```python
from matplotlib.colors import ListedColormap
```

Функция scatter позволяет отрисовывать графики на входе:
x - координаты, y - координаты

используем функцию map для того, чтобы получить нужные столбцы из массива данных.
map(lambda x: x[0], circles[0]) - получаем 0 стобец 0 столбца.

Передаем массив признаков и массив цветов для признаков
colors = ListedColormap(['red', 'yellow'])
c = circles[1], cmap = colors

```python
colors = ListedColormap(['red', 'yellow'])

pyplot.figure(figsize(8, 8))
pyplot.scatter(map(lambda x: x[0], circles[0]), map(lambda x: x[1], circles[0]), c = circles[1], cmap = colors)
```

```python
def plot_2d_dataset(data, colors):
    pyplot.figure(figsize(8, 8))
    pyplot.scatter(map(lambda x: x[0], data[0]), map(lambda x: x[1], data[0]), c = data[1], cmap = colors)
```

Добавляем шум к наборку данных и передаем его в функцию отрисовки графика

```python
noisy_circles = datasets.make_circles(noise = 0.15)
```

```python
plot_2d_dataset(noisy_circles, colors)
```

#### datasets.make_classification
Создание своего набора данных
 - n_features = 2, - количество признаков
 - n_informative = 1, - информативные признаки
 - n_redundant = 1, - избыточные признаки
 - n_clusters_per_class = 1, - количество кластеров в классе
 - random_state = 1

```python
simple_classification_problem = datasets.make_classification(n_features = 2, n_informative = 1, 
                                                            n_redundant = 1, n_clusters_per_class = 1,
                                                            random_state = 1 )
```

```python
plot_2d_dataset(simple_classification_problem, colors)
```

```python
classification_problem = datasets.make_classification(n_features = 2, n_informative = 2, n_classes = 4, 
                                                      n_redundant = 0, n_clusters_per_class = 1, random_state = 1)

colors = ListedColormap(['red', 'blue', 'green', 'yellow'])
```

```python
plot_2d_dataset(classification_problem, colors)
```

### Визуализация выборки

```python
from pandas import DataFrame
```

```python
iris_frame = DataFrame(iris.data)
# добавим к объекту имена классов
iris_frame.columns = iris.feature_names
# добавим столбец с метками классов
iris_frame['target'] = iris.target
```

```python
iris_frame.head()
```

DataFrame.column.apply(lambda x : iris.target_names[x])
производит операцию для всех значений столбца, в зависимости от значения подставляем индексы имен меток классов

```python
iris_frame.target = iris_frame.target.apply(lambda x : iris.target_names[x])
```

```python
iris_frame.head()
```

Выбираем данные из столбца с определенным признаком и строим гистограмму распределения.

```python
iris_frame[iris_frame.target == 'setosa'].hist('sepal length (cm)')
```

```python
# устнавливаем размер графика
pyplot.figure(figsize(20, 24))

plot_number = 0
# в цикле перебираем имена признаков
for feature_name in iris['feature_names']:
    # а в нем имена меток
    for target_name in iris['target_names']:
        plot_number += 1
        # устанавливаем размерность матрицы и 
        # передаем индекс текущего графика
        pyplot.subplot(4, 3, plot_number)
        # устанавливаем текущие значение признаков и меток
        pyplot.hist(iris_frame[iris_frame.target == target_name][feature_name])
        pyplot.title(target_name)
        pyplot.xlabel('cm')
        pyplot.ylabel(feature_name[:-4])
        # можно установить одинаковые размерности координатной сетки
        # pyplot.xlim([0,10])
        # pyplot.ylim([0,30])
```
